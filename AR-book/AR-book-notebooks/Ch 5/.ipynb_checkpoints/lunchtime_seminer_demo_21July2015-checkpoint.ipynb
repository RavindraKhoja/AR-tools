## Setup

```python
import scipy as sp
import matplotlib.pyplot as plt

from scipy.integrate import odeint
from scipy.spatial import ConvexHull
import scipy.optimize

from IPython.html.widgets import interact, interactive, fixed, Checkbox
from IPython.display import display

%matplotlib inline

#2D van de Vusse kinetics
k1 = 1.0
k2 = 1.0
k3 = 10.0
def rate_fn(C,t):
    
    cA = C[0]
    cB = C[1]

    #r = [rA, rB]
    return sp.array([-k1*cA - 2*k3*cA**2,
                    k1*cA - k2*cB])

#feed
Cf = sp.array([1., 0.])

#integrate PFR
t_end = 10;
N_pts = 50;
pfr_ts = sp.logspace(-3,sp.log10(t_end),N_pts)

pfr_cs = odeint(rate_fn,Cf,pfr_ts)

#solve CSTR
cstr_ts = sp.logspace(-3,sp.log10(t_end*100),N_pts)

k=0
C_guess = Cf
cstr_cs = Cf
for ti in cstr_ts:
    def cstr_fn(Ci):
        f = Cf + ti*rate_fn(Ci,[]) - Ci
        
        return f
    
    #attempt to solve for CSTR solution
    Ci = scipy.optimize.newton_krylov(cstr_fn, C_guess, f_tol=1e-4)
    cstr_cs = sp.vstack([cstr_cs, Ci])
    C_guess = Ci
    
    k += 1
    
#PFR from CSTR locus
cstrpfr_cs = odeint(rate_fn,cstr_cs[25,:],pfr_ts)

#combine all points
all_cs = sp.vstack([cstr_cs,cstrpfr_cs])

K = ConvexHull(all_cs).vertices
chull_xs, chull_ys = all_cs[K,0], all_cs[K,1]
```

# AR Construction demonstration

```python
def plot_fn(plot_pfr=False, plot_cstr=False, plot_cstrpfr=False, plot_region=False):
    
    region_cs = []
    
    plt.figure()
    plt.hold(True)
        
    if plot_pfr:
        plt.plot(pfr_cs[:,0],pfr_cs[:,1],'r-')
        region_cs = pfr_cs
        
    if plot_cstr:
        plt.plot(cstr_cs[:,0],cstr_cs[:,1], 'bx')
        region_cs = sp.vstack([region_cs, cstr_cs])
        
        if plot_cstrpfr:
            plt.plot(cstrpfr_cs[:,0],cstrpfr_cs[:,1], "r-")
            region_cs = sp.vstack([region_cs, cstrpfr_cs])
        
    if plot_region:
        K = ConvexHull(region_cs).vertices
        chull_xs, chull_ys = region_cs[K,0], region_cs[K,1]
        plt.fill(chull_xs, chull_ys, 'g', alpha=0.3)
        
    plt.axis([0,1,0,0.12])
    plt.xlabel('cA (mol/L)')
    plt.ylabel('cB (mol/L)')

    plt.show()
    
interactive(plot_fn)
```

##Profit objective function

Suppose that the profit of the process is a function of the following form:

$$ P=c_{B}-Sc_{A} $$

B is the desired product and A is an undesired reactant that musy be separated.

```python
def plot_fn2(S=0.25, P=0.05):
      
    def objective_fn(cA):
        return cA*S + P
    
    obj_cAs = sp.linspace(0,1,10)
    obj_cBs = objective_fn(obj_cAs)
    
    plt.figure()
    plt.hold(True)

    plt.fill(chull_xs, chull_ys, 'g', alpha=0.3)
    plt.plot(pfr_cs[:,0],pfr_cs[:,1],'r-')
    plt.plot(cstr_cs[:,0],cstr_cs[:,1], 'bx')
    
    plt.plot(obj_cAs,obj_cBs,'k:')

    plt.title('cB = %.2fcA + %.3f mol/L'%(S,P))
    plt.axis([0,1, 0,0.14])
    plt.xlabel('cA (mol/L)')
    plt.ylabel('cB (mol/L)')
    plt.show()
    
interactive(plot_fn2, S=(0,1,0.02), P=(0,0.12,0.005))
```

##Yield objective function

We use instantaneous yield of component B over component D as an objective function.

$$Y_{BD} = \frac{k_1c_A - k_2c_B}{k_3{c_A}^2}$$

When plotted against the AR, this provides insight into which concentrations should be targeted to favour the production of either component B or D.

```python
def plot_fn3(Y_BD=0.2):
      
    def objective_fn(cA):
        return (k1*cA - Y_BD*k3*cA**2)/k2
    
    obj_cAs = sp.linspace(0,1,50)
    obj_cBs = objective_fn(obj_cAs)
    
    plt.figure()
    plt.hold(True)

    plt.fill(chull_xs, chull_ys, 'g', alpha=0.3)
    plt.plot(pfr_cs[:,0],pfr_cs[:,1],'r-')
    plt.plot(cstr_cs[:,0],cstr_cs[:,1], 'bx')
    
    plt.plot(obj_cAs,obj_cBs,'k:')

    plt.title('Yield $Y_{BD}$ = %.3f'%(Y_BD))
    plt.axis([0,1, 0,0.14])
    plt.xlabel('cA (mol/L)')
    plt.ylabel('cB (mol/L)')
    plt.show()
    
interactive(plot_fn3, Y_BD=(0.1,0.5,0.001))
```
